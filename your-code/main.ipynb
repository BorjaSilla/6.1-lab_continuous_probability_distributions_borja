## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
from scipy.stats import uniform
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
from scipy.stats import uniform
x = uniform.rvs(size=10)
a = 2
b = 3
randoms  = a + (b-a)*x
print(randoms)
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
def generate_uniform_random_numbers(bottom, ceiling, count):
    x = uniform.rvs(size=count)
    randoms = bottom + (ceiling - bottom) * x
    return randoms

# Call the function with the specified parameters
random_numbers_1 = generate_uniform_random_numbers(10, 15, 100)
random_numbers_2 = generate_uniform_random_numbers(10, 60, 1000)

# Plot histograms with shared y-axis
fig, axs = plt.subplots(1, 2, figsize=(12, 6), sharey=True)

axs[0].hist(random_numbers_1, bins=10)
axs[0].set_title('Histogram - Random Numbers (10, 15, 100)')
axs[0].set_xlabel('Value')
axs[0].set_ylabel('Count')

axs[1].hist(random_numbers_2, bins=10)
axs[1].set_title('Histogram - Random Numbers (10, 60, 1000)')
axs[1].set_xlabel('Value')

plt.tight_layout()
plt.show()
```

How are the two distributions different?

```python
# The ceiling, or upper boundary in dist 2 is 45 units higher and the count is 10 times greater. (100 vs 1000)
```

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_normal_random_numbers(mean, std_dev, count):
    random_numbers = np.random.normal(mean, std_dev, count)
    return random_numbers

# Generate normally distributed numbers with mean=10 and std_dev=1
normal_numbers_1 = generate_normal_random_numbers(10, 1, 1000)

# Generate normally distributed numbers with mean=10 and std_dev=50
normal_numbers_2 = generate_normal_random_numbers(10, 50, 1000)

# Plot the distributions
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(normal_numbers_1, bins=30)
plt.title('Normal Distribution (Mean=10, Std Dev=1)')
plt.xlabel('Value')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
plt.hist(normal_numbers_2, bins=30)
plt.title('Normal Distribution (Mean=10, Std Dev=50)')
plt.xlabel('Value')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

How are the two distributions different?

```python
# the key difference is in the spread or dispersion of values around the mean: 
# the first distribution is more concentrated, while the second distribution has a broader range of values.
```

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e.  = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
import math

def exponential_cdf(x, lmbda):
    """
    Calculate the cumulative distribution function (CDF) of the exponential distribution.
    """
    return 1 - math.exp(-lmbda * x)

# Given mean is 10 minutes, so lambda = 1/10
lmbda = 1/10

# Time threshold is 15 minutes
time_threshold = 15

# Calculate the probability that a customer will spend less than 15 minutes
probability_less_than_15_minutes = exponential_cdf(time_threshold, lmbda)

print(f"The probability that a customer will spend less than fifteen minutes in the bank is: {probability_less_than_15_minutes:.4f}")
```

What is the probability that the customer will spend more than 15 minutes

```python
# Given mean is 10 minutes, so lambda = 1/10
lmbda = 1/10

# Time threshold is 15 minutes
time_threshold = 15

# Calculate the probability that a customer will spend more than 15 minutes
probability_more_than_15_minutes = 1 - exponential_cdf(time_threshold, lmbda)

print(f"The probability that a customer will spend less than fifteen minutes in the bank is: {probability_more_than_15_minutes:.4f}")
```
